```python
from pprint import pprint
from paddlenlp import Taskflow
```

```python
insight = 'The Utah Jazz have helped Lauri Markkanen unlock his offensive potential, becoming the kind of star they build a new foundation around.'
#insight = "The Warriors trailed by six with three minutes left before Durant, criticized for leaving Oklahoma City last summer to chase a championship, brought them back, scoring 14 in the fourth."
```

```python
# Named Entity Recognition (NER)
ner = Taskflow("ner")
pprint(ner(insight))
```

```python
ner_fast = Taskflow("ner", mode="fast")
pprint(ner_fast(insight))
```

```python
schema = ['Person', 'Team', 'Place']
ie_en = Taskflow('information_extraction', schema=schema, model='uie-base-en')
pprint(ie_en(insight))
```
